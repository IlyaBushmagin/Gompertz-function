```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
f = pd.read_csv('btc.csv')
f
```

```python
plt.figure(figsize=(8, 8))
plt.plot(f['Close'])
plt.show()
```

```python
F = []
F.append(f['Close'][0])
for i in range(len(f) - 1):
	F.append(F[i] + f['Close'][i + 1])
```

```python
plt.figure(figsize=(8, 8))
plt.plot(F)
plt.show()
```

```python
x = np.array(range(15, len(f)))
y = np.log(f['Close'][15:] / F[15:])
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
```

```python
plt.figure(figsize=(8, 8))
plt.plot(x, p(x))
plt.plot(x, y, 'rx')
plt.show()
```
